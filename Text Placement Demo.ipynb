```python
import pandas as pd
from IPython.core.display import HTML
```

```python
from svgtext import *
```

```python
df = pd.read_csv("enterprise_app_data.csv")
name_dict = dict(df[['System', 'Description']].values)
```

```python
w,h = 120,25
xx,yy = 600,300
rects = []
columns = 7

bw = xx/(columns+1)
bh = 30
for e,(k,v) in enumerate(name_dict.items()):
    x = (e%columns)*(xx/columns)
    y = (e//columns) * (yy/columns)
    if e % 5 == 0:
        rects.append(titled_panel(x, y+(bh/5), bw, bh+(bh/5), 5, 3*(bh/5), v, oversize_method="wrap", text_align="centre", debug=True ))
    else:
        rects.append(text_panel(x, y+(bh/5), bw, bh+(bh/5), 5, v, oversize_method="wrap", text_align="centre", debug=True ))


        
svg_objects = "".join(rects)
content = svg_rect(-20,-20,xx+60,yy) + svg_objects
HTML(svg_viewbox(-20,-20,800,800,600,500, content))
```
